```python
import re

with open("row.txt", 'r', encoding = "utf-8") as file:
    data = file.read()
print("Task 1")

pattern = r"ab*"

for match in data.split('\n'):
    matches = re.findall(pattern, match)
    if matches:
        print(f'{match} = {matches}')
```

```python
import re


def read_file(filepath):
    with open(filepath, 'r', encoding='utf-8') as file:
        return file.read()


file_path = "row.txt"  
text = read_file(file_path)

# 1. 'a' символын және одан кейінгі 'b' символдарын табу
pattern1 = re.findall(r'a*b*', text)
print("1:", pattern1)

# 2. 'a' символынан кейін 2 немесе 3 'b' символдары бар жолдарды табу
pattern2 = re.findall(r'ab{2,3}', text)
print("2:", pattern2)

# 3. Кіші әріптер мен төменгі сызықтардан тұратын тізбектерді табу
pattern3 = re.findall(r'\b[a-z]+_[a-z]+\b', text)
print("3:", pattern3)

# 4. Бір бас әріптен басталатын және одан кейін кіші әріптер келетін сөздерді табу
pattern4 = re.findall(r'\b[A-Z][a-z]+\b', text)
print("4:", pattern4)

# 5. 'a' символынан басталып, кез келген таңбалар арқылы 'b' символымен аяқталатын жолдарды табу
pattern5 = re.findall(r'a.*b', text)
print("5:", pattern5)

# 6. Барлық бос орындарды, үтірлерді және нүктелерді қос нүктемен ауыстыру
text_replaced = re.sub(r'[ ,.]', ':', text)
print("6:", text_replaced)

# 7. snake_case мәтінін CamelCase-ке түрлендіру
snake_case_text = "hello_world_example"
camel_case_text = ''.join(word.capitalize() for word in snake_case_text.split('_'))
print("7:", camel_case_text)

# 8. Жолды бас әріптер бойынша бөлу (алғашқы 10 элемент)
pattern8 = re.split(r'(?=[A-Z])', text)
print("8:", pattern8[:10])  # Тек алғашқы 10 элементті көрсетеміз

# 9. Бас әріптерден басталатын сөздердің арасына бос орын енгізу (алғашқы 500 символ)
pattern9 = re.sub(r'([a-z])([A-Z])', r'\1 \2', text)
print("9:", pattern9[:500])  # Тек алғашқы 500 символды көрсетеміз

# 10. CamelCase мәтінін snake_case-ке түрлендіру
camel_case_text = "HelloWorldExample"
snake_case_text = re.sub(r'([a-z])([A-Z])', r'\1_\2', camel_case_text).lower()
print("10:", snake_case_text)


```
